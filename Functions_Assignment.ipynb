```python
# Q.4 Write a Python function that checks if a given number is prime or not from 1 to 200.
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_numbers = [num for num in range(1, 200) if is_prime(num)]
```

```python
print(prime_numbers)
```

```python
is_prime(5)
```

```python
is_prime(42)
```

```python

def read_file_line_by_line(file_path):
    """
    Generator function that reads a file line by line and yields each line as a string.

    :param file_path: Path to the file to be read.
    :yield: Each line in the file as a string.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Use rstrip() to remove the newline character
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except IOError:
        print(f"An error occurred while reading the file: {file_path}")
```

```python
file_path = 'goodbye'
for line in read_file_line_by_line(file_path):
    print(line)
```

```python
# Q.11 Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:


#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,-  if the value of the order is smaller than 100,00 .

#Write a Python program using lambda and map.
```

```python
book_orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfhrung in Python3, Bernd Klein", 3, 24.99]
]

```

```python
calcu_total = lambda order: (order[0], order[2] * order[3] if order[2] * order[3] >= 100 else order[2] * order[3] + 10)

order_totals = list(map(calcu_total, book_orders))

print(order_totals)
```
